```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load historical data
df = pd.read_csv('/content/options_and_futures_risk_management.csv')
```

```python
# Define trading strategy
def moving_average_strategy(df, short_window, long_window):
    df['Signal'] = np.where(df['Underlying_Price'] > df['Underlying_Price'].rolling(window=short_window).mean(), 1, 0)
    df['Position'] = df['Signal'].diff()
    return df
```

```python
# Backtest strategy
df = moving_average_strategy(df, 50, 200)

# Calculate strategy returns
df['Strategy_Returns'] = df['Underlying_Price'].pct_change() * df['Position'].shift(1)

# Evaluate strategy performance
sharpe_ratio = df['Strategy_Returns'].mean() / df['Strategy_Returns'].std() * np.sqrt(252)
sortino_ratio = df['Strategy_Returns'].mean() / df['Strategy_Returns'][df['Strategy_Returns'] < 0].std() * np.sqrt(252)
print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
print(f"Sortino Ratio: {sortino_ratio:.2f}")

# Plot strategy performance
plt.plot(df['Underlying_Price'], label='Underlying Price')
plt.plot(df['Underlying_Price'] * df['Position'], label='Strategy')
plt.legend()
plt.show()
```
